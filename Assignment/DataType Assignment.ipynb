```python
# You are given a list of integers. You are asked to make a list by running

	# i.e. Say given list is [2,4,6,10,1] resultant list will be [22,20,10,23].

	# For 1st element 2 ->> these are greater (4+6+10) values and 2 itself so on adding becomes 22.

	# For 2nd element 4 ->> greater elements are (6, 10) and 4 itself, so on adding 20

	# like wise for all other elememts.
list = [2,4,6,10,1]
result = []
for i in list:
    total = i
    for j in list:
        if j > i:
            total = total+j
    result.append(total)
print(result)
```

```python
# Sort a list of alphanumeric strings based on product value of numeric character in it. If in any string there is no numeric character take it's product value as 1.
list = ['1ac21', '23fg', '456', '098d','1','kls']
def product_of_digits(s):
    product = 1
    has_digit = False

    for ch in s:
        if ch.isdigit():
            product = product*int(ch)
            has_digit = True

    return product if has_digit else 1
    #sort based on product
sorted_list = sorted(list, key=product_of_digits,reverse =True)
print(sorted_list)
            
```

```python
# Write a Python function to check whether a number is perfect or not.¶
# A Perfect number is a number that is half the sum of all of its positive divisors (including itself).
def is_perfectnumber(n):
    total =0
    for i in range(1,n):
        #check divisior of n
        if n % i ==0:
            total = total + i
            
    if total == n:
        return True
    else:
        return False
print(is_perfectnumber(28))
            
            
            
    
```

```python
# Write a python function that receives a list of integers and prints out a histogram of bin size 10
def histogram_bins(lst):
    bins = {}

    for num in lst:
        #Find bin range
        lower = (num //10)*10 + 1
        upper = lower+9

        key = f"{lower}-{upper}"

        #count frequency
        if key in bins :
            bins[key]+=1
        else:
            bins[key] = 1
    # sort bins by lower range
    sorted_bins = dict(sorted(bins.items(),key = lambda x: int(x[0].split('-')[0])))
    print(sorted_bins)
histogram_bins([13,42,15,37,22,39,41,50])
```
