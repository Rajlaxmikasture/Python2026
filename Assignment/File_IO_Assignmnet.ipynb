```python
import os
text_to_add = 'welcome to my page'

if os.path.exists('sample.txt'):
  with open('sample.txt','r') as f:
    content = f.read()
else:
  content=""

if text_to_add not in content:
  with open('sample.txt','a') as f:
    f.write("\n"+text_to_add)

with open('sample.txt','r') as f:
  print(f.read())
```

```python
text_to_remove = "welcome to my page "

#read all lines
with open('sample.txt','r') as f:
  lines = f.readlines()

#Remove the matching line
with open('sample.txt','w')as f:
  for line in lines:
    if line.strip()!=text_to_remove:
      f.write(line)

#Display updated content
with open('sample.txt','r')as f:
  print(f.read())
```

```python
# Write a Python program to count the frequency of words in a file.
#using dictionary
with open('sample.txt','r') as f:
  word_count = {}

  for line in f:
    words = line.strip().split() #split line into words

    for word in words:
      if word in word_count:
        word_count[word]+=1
      else:
        word_count[word]=1
    print("word frequency:")
    for word,count in word_count.items():
      print(word ,":", count)
```

```python
#Using collections
from collections import counter
with open('sample.txt','r')as f:
   words = f.read().lower().split()
word_count = Counter(words)
print(word_count)
```
