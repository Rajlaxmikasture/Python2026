```python
# Write a Python function to sum all the numbers in a list.
def sum(l):
    sum = 0
    for i in l:
         sum = sum+i
    return sum
l = (8, 2, 3, 0, 7)
print(sum(l))
```

```python
# Write a Python function to multiply all the numbers in a list.
def multiply(l):
    count = 1
    for i in l:
        count = count * i
    return count
l =[8, 2, 3, -1, 7]
print(multiply(l))
```

```python
# Write a Python function that 
def count_case_letters(s):
    upper_count = 0
    lower_count = 0

    for ch in s:
        if ch.isupper():
            upper_count +=1
        elif ch.islower():
            lower_count+=1
    print("Uppercase letters:" ,upper_count)
    print("Lowercase letters:",lower_count)

#Example usage
count_case_letters('The quick Brow Fox')
            
```

```python
# Write a Python function that takes a list and returns a new list with distinct elements from the first list.
def unique_list(l):
    unique_list = []
    for i in l:
        if i not in unique_list:
            unique_list.append(i)
    return unique_list
l = [1,2,3,3,3,3,4,5]
print(unique_list(l))
```

```python
# Write a Python function that checks whether a passed string is a palindrome or not.
def ispalindrom_string(s):
    rev = ""
    for ch in s:
        rev = ch + rev 
   
    if rev == s :
        return "palindrome string"
    else:
        return "Not a plaindrome" 
print(ispalindrom_string('madam'))
        
```

```python
 # Write a Python program that accepts a hyphen-separated sequence of words as input and prints the words in a hyphen-separated sequence after sorting them alphabetically.
def sort_hyphen_word(s):
    words = s.split('-')  #split into list
    words.sort()          #sort alphabetically
    return '-'.join(words) #join back with hyphen

#example
input_str = "green-red-yellow-black-white"
print(sort_hyphen_word(input_str))
```

```python
# Write a Python program to create a function that takes one argument, and that argument will be multiplied with an unknown given number.
#The function should return a new function that multiplies n with that unknown value
#This is done using a closure(function inside a function)
def my_function(n):
    return lambda x : x * n

#create a function that multiplies by 2
result = my_function(2)
result2 = my_function(3)
result3 = my_function(4)
result4 = my_function(5)

print("Double of the number of 15 = ",result(15))
print("Triple the number of 15 =",result2(15))
print("Quadruple the number of 15 =",result3(15))
print("Quintuple the number 15 = ",result4(15))
```

```python
# Write a Python program to count the even and odd numbers in a given array of integers using Lambda.
numbers = [1, 2, 3, 5, 7, 8, 9, 10]
count_even_number = sum( 1 for x in numbers if x%2==0)
print("Numbers of even numbers in the above array :",count_even_number)
count_odd_numbers = sum(1 for x in numbers if x%2 !=0)
print("Numbers of odd numbers in the above array :",count_odd_numbers)

# Function needs two argument filter(function ,iterable)
```

```python
# Write a Python program to check whether a given string contains a capital letter, a lower case letter, a number and a minimum length using lambda.
s = input("Enter a string : ")

#conditions using lambda
has_upper = lambda x:any(ch.isupper() for ch in x)
has_lower = lambda x: any(ch.islower() for ch in x)
has_digit = lambda x : any(ch.isdigit() for ch in x)
min_length = lambda x : len(x) >=8

#check all conditions
if has_upper(s) and has_lower(s) and has_digit(s) and min_length(s) :
    print("vaild string")
else:
    print("Invaild string")
```
